```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### AEP

```python
out_dir = 'AEP_results'
result5 = pd.read_csv('new_exp_results/' + out_dir + '/results_fourier.csv')
```

```python
import seaborn as sns

g = sns.FacetGrid(result5, col="hour", col_wrap=6, height=2.5, sharex=True, sharey=True)
g.map_dataframe(sns.histplot, x="p_values", bins=10, kde=False, color="skyblue")

g.set_titles("h_ahead = {col_name}")
g.set_axis_labels("p-value", "Count")
g.tight_layout()
plt.suptitle("Histogram of p-values by h_ahead", y=1.02)
plt.savefig('new_exp_results/' + out_dir + '/p_values_plot_aep.png')
plt.show()
```

```python
result5['rejected'] = ((result5['y'] < result5['y_hat_lower']) | (result5['y'] > result5['y_hat_upper'])).astype(int)
rejection_by_hour = result5.groupby('hour').agg({'rejected': ['sum', 'count', 'mean']}).reset_index()
rejection_by_hour.columns = ['hour', 'n_rejected', 'n_total', 'rejection_rate']
print(rejection_by_hour)
```
